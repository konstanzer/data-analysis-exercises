```python
import pandas as pd
import numpy as np

path = "/content/drive/MyDrive/Colab Notebooks/data/untidy-data/"

att = pd.read_csv(path + "attendance.csv")
cof = pd.read_csv(path + "coffee_levels.csv")
cak = pd.read_csv(path + "cake_recipes.csv")
```

```python
att=att.rename(columns={"Unnamed: 0":"who"})
att.info()
```

```python
att=att.melt(id_vars='who', var_name='when', value_name='attendance')
att=att.replace({"H":.5, "T":.9, "A": 0, "P":1})
att
```

```python
att=att.groupby("who").mean()
att
```

```python
cof.pivot(columns="coffee_carafe", index="hour")
```

```python
cak=pd.concat([cak["recipe:position"].str.split(":", expand=True), cak], axis=1).drop("recipe:position", axis=1)
cak=cak.rename(columns={0:"recipe", 1:"rack"})
cak=cak.melt(id_vars=["recipe", "rack"], var_name="temperature", value_name="tastiness")
cak
```

```python
cak.groupby("recipe").mean() #b wins
```

```python
cak.groupby("temperature").mean() #275 wins
```
